## 'fuel_type'

```python
# Fill missing 'fuel_type' values when the model is 'Tesla'
test.loc[(test['fuel_type'].isnull()) & (test['brand'] == 'Tesla'), 'fuel_type'] = 'Electric'

print(test['fuel_type'].isnull().sum())
```

```python
# Use str.contains to find rows where 'engine' contains 'Electric' (case-insensitive)
# Only affect rows where 'fuel_type' is null
test.loc[(test['fuel_type'].isnull()) & (test['engine'].str.contains('Electric', case=False, na=False)), 'fuel_type'] = 'Electric'

print(test['fuel_type'].isnull().sum())
```

```python
# Filter and display rows where 'fuel_type' is still null
null_fuel_type_rows = test[test['fuel_type'].isnull()]
null_fuel_type_rows.head(79)
```

```python
# Use str.contains to find rows where 'engine' contains 'Gasoline' (case-insensitive)
# Only affect rows where 'fuel_type' is null
test.loc[(test['fuel_type'].isnull()) & (test['engine'].str.contains('Gasoline', case=False, na=False)), 'fuel_type'] = 'Gasoline'

print(test['fuel_type'].isnull().sum())
```

```python
# Filter and display rows where 'fuel_type' is still null
null_fuel_type_rows = test[test['fuel_type'].isnull()]
null_fuel_type_rows.head(45)
```

```python
# Typically, the terms "Dual Motor" and 'Standard Range Battery'refer to electric vehicles

# Fill 'fuel_type' with 'Electric' for cars with 'Dual Motor - Standard' or 'Standard Range Battery' in 'engine'
test.loc[(test['fuel_type'].isnull()) & (test['engine'].str.contains('Dual Motor - Standard|Standard Range Battery', case=False, na=False)), 'fuel_type'] = 'Electric'


print(test['fuel_type'].isnull().sum())
```

```python
# Filter and display rows where 'fuel_type' is still null
null_fuel_type_rows = test[test['fuel_type'].isnull()]
null_fuel_type_rows
```

```python
# Given the term "GDI" (Gasoline Direct Injection), it is safe to assume that engines with "GDI DOHC" are gasoline-powered vehicles.

# Fill 'fuel_type' with 'Gasoline' for cars with 'GDI DOHC' in 'engine' and where 'fuel_type' is null
test.loc[(test['fuel_type'].isnull()) & (test['engine'].str.contains('GDI DOHC', case=False, na=False)), 'fuel_type'] = 'Gasoline'


print(test['fuel_type'].isnull().sum())
```

```python
# Filter and display rows where 'fuel_type' is still null
null_fuel_type_rows = test[test['fuel_type'].isnull()]
null_fuel_type_rows
```

```python
# Fill 'fuel_type' for electric vehicles
test.loc[(test['fuel_type'].isnull()) & (test['engine'].str.contains('120 AH|F-150 Lightning|Q4 e-tron|EV6', case=False, na=False)), 'fuel_type'] = 'Electric'

# Fill 'fuel_type' for diesel engines
test.loc[(test['fuel_type'].isnull()) & (test['engine'].str.contains('Turbo Diesel', case=False, na=False)), 'fuel_type'] = 'Diesel'

# Fill 'fuel_type' for gasoline turbo engines
test.loc[(test['fuel_type'].isnull()) & (test['engine'].str.contains('Twin Turbo', case=False, na=False)), 'fuel_type'] = 'Gasoline'


print(test['fuel_type'].isnull().sum())
```

```python
# Check remaining null values in 'fuel_type'
null_fuel_type_rows = test[test['fuel_type'].isnull()]
null_fuel_type_rows
```

```python
# Fill 'fuel_type' for electric vehicles (F-150 Lightning, Kia EV6, Audi Q4 e-tron)
test.loc[(test['fuel_type'].isnull()) & (test['model'].str.contains('F-150 Lightning|EV6|Q4 e-tron', case=False, na=False)), 'fuel_type'] = 'Electric'

# Fill 'fuel_type' for gasoline vehicles (Nissan Kicks, Mercedes-Benz S-Class, Ford Air Pure, Audi Q5, Mercedes-Benz R-Class)
test.loc[(test['fuel_type'].isnull()) & (test['model'].str.contains('Kicks|S-Class|Air Pure|Q5|R 350', case=False, na=False)), 'fuel_type'] = 'Gasoline'
```

```python
# Check remaining null values in 'fuel_type'
print(test['fuel_type'].isnull().sum())
```

```python
# removing spaces from the 'engine' column to avoid mismatches
train['engine'] = train['engine'].str.strip()

# Fill 'fuel_type' with 'Electric' for electric vehicles
train.loc[(train['fuel_type'].isnull()) & (train['engine'].str.contains('Dual Motor - Standard|Standard Range Battery|electric', case=False, na=False)), 'fuel_type'] = 'Electric'

print(train['fuel_type'].isnull().sum())
```

```python
# Fill 'fuel_type' for BMW i3 with Range Extender as Hybrid
train.loc[(train['fuel_type'].isnull()) & (train['model'].str.contains('i3 120Ah w/Range Extender', case=False, na=False)), 'fuel_type'] = 'Hybrid'

print(train['fuel_type'].isnull().sum())
```

```python
# Fill 'fuel_type' for gasoline engines
train.loc[(train['fuel_type'].isnull()) & (train['engine'].str.contains('GDI DOHC|Twin Turbo', case=False, na=False)), 'fuel_type'] = 'Gasoline'

print(train['fuel_type'].isnull().sum())
```

```python
# Fill 'fuel_type' for diesel engines
train.loc[(train['fuel_type'].isnull()) & (train['engine'].str.contains('Turbo Diesel', case=False, na=False)), 'fuel_type'] = 'Diesel'

print(train['fuel_type'].isnull().sum())
```
